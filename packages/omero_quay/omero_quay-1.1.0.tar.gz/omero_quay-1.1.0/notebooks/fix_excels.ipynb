```python
import os
from pathlib import Path

import pandas as pd
```

```python

excels_path = Path(os.environ.get('QUAY_TEST_DATA')) / 'excels' / 'validation_excels'


prov = pd.DataFrame(columns=['provenance', 'destination'])
prov.loc[0] = ['local', 'docker']

for xlsx in excels_path.rglob("*.xlsx"):
    print(xlsx)
    try:
        usr = pd.read_excel(xlsx, 'User')
    except:
        usr = None
    try:
        inv  = pd.read_excel(xlsx, 'Investigation')
        if not 'provenance' in inv.columns:
            inv  = inv.drop('samba_path', axis=1)
            inv = pd.concat([inv, prov], axis=1)
        elif inv.loc[0, "provenance"] == 'docker':
            inv["provenance"] = "local"
            print(inv["provenance"])
    except:
        inv = None
    
    try:
        stu  = pd.read_excel(xlsx, 'Study')
        stu.columns = ['owner', 'name', 'description', 'parent', 'tags']
    except:
        stu = None
    try:
        ass  = pd.read_excel(xlsx, 'Assay')
        ass['path'] = ass['path'].apply(lambda s: f'S-JCBD-200902093/{s}')
    except ValueError:
        ass = None
    
    with pd.ExcelWriter(xlsx) as dest:
        if usr is not None:
            usr.to_excel(dest, sheet_name='User', index=False)
        if inv is not None:
            inv.to_excel(dest, sheet_name='Investigation', index=False)
        if stu is not None:
            stu.to_excel(dest, sheet_name='Study', index=False)
        if ass is not None:
            ass.to_excel(dest, sheet_name='Assay', index=False)
    
```
