# Basic Usage of Boxscore subpackage.

```python
from dans.endpoints.boxscore.bxplayerlogs import BXPlayerLogs
from dans.endpoints.boxscore.bxplayerstats import BXPlayerStats
from dans.library.parameters import SeasonType, DataFormat
```

```python
logs = BXPlayerLogs("Kevin Durant", [2019, 2021], season_type=SeasonType.playoffs).bball_ref()
logs
```

Kevin Durant did not finish two games during this stretch due to injury (G5 vs. Houston and G5 @ Toronto). Let's remove these games.

```python
logs_to_drop = logs.iloc[[10, 11]].index
logs = logs.drop(logs_to_drop)

stats = BXPlayerStats(player_logs=logs, drtg_range=[80, 130], data_format=DataFormat.per_game).bball_ref()
stats
```

```python
stats = BXPlayerStats(player_logs=logs, drtg_range=[80, 130], data_format=DataFormat.per_100_poss).bball_ref()
stats
```

Notice that the `DRTG` column changed. This is because the average defensive strength of a player's opponents is weighted by games when using `per_game`, whereas it is weighted by possessions when using `per_100_poss` (and other data formats that involve possessions or pace) to improve accuracy. This applies to all other opponent defensive metrics, such ass `OPP_TS`.

```python
stats = BXPlayerStats(player_logs=logs, drtg_range=[80, 130], data_format=DataFormat.pace_adj).bball_ref()
stats
```

```python
stats = BXPlayerStats(player_logs=logs, drtg_range=[80, 130], data_format=DataFormat.opp_adj).bball_ref()
stats
```

```python
stats = BXPlayerStats(player_logs=logs, drtg_range=[80, 130], data_format=DataFormat.opp_pace_adj).bball_ref()
stats
```
